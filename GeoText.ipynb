```python
import os
import re
from geotext import GeoText

txt_directory = "/Users/macbookpro/Desktop/Scraping/FMLN-txt"
output_file_path = "/Users/macbookpro/Desktop/Scraping/country_mentions.txt"

# Keywords to identify in the text
keywords = ["España", "Perú", "Inglaterra", "Estados Unidos", "USSR", "Unión Soviética", "Rusia"]

with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for filename in os.listdir(txt_directory):
        if filename.endswith(".txt"):
            txt_file_path = os.path.join(txt_directory, filename)

            with open(txt_file_path, 'r', encoding='utf-8') as file:
                sentence = file.read()

            places = GeoText(sentence)
            #print(f"File: {filename}")
            #print("Countries:", places.countries)
            output_file.write(f"File: {filename}\n")
            output_file.write(f"Countries: {places.countries}\n")

            for keyword in keywords: 
                keyword_lower = keyword.lower()
                occurrences = [match.start() for match in re.finditer(keyword_lower, sentence.lower())]

                if occurrences:
                    for occurence in occurrences: 
                        #print(f"{keyword}")
                        output_file.write(f"Other Countries: {keyword}\n")
```

```python
from collections import Counter
import re

file_path = "Desktop/Scraping/country_mentions.txt"

with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

words = re.findall(r'\b\w+\b', content)

word_frequency = Counter(words)

sorted_words = word_frequency.most_common()

for word, frequency in sorted_words:
    print(f"{word}: {frequency} times")
```

```python
import os
import pandas as pd

txt_directory = "/Users/macbookpro/Desktop/Scraping/Peru-SL-txt"

dfs = []

for filename in os.listdir(txt_directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(txt_directory, filename)
        
        document_title = os.path.splitext(filename)[0]

        
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

        china_occurrences = content.lower().count('mao')


        df = pd.DataFrame({'File Name': [document_title], 'Mao Occurrences': [china_occurrences]})

        dfs.append(df)


result_df = pd.concat(dfs, ignore_index=True)


print(result_df)

#result_df.to_csv('china_occurrences.csv', index=False)

```

```python
#merging the data frames 
import pandas as pd
shiningPathdf['File Name'] = shiningPathdf['File Name'].astype(str)
result_df['File Name'] = result_df['File Name'].astype(str)
merged_df = pd.merge(shiningPathdf, result_df, on='File Name', how='inner')

print(merged_df)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt


merged_df['Date'] = pd.to_datetime(merged_df['Date'])


merged_df['Period'] = merged_df['Date'].dt.year // 5 * 5

period_totals = merged_df.groupby('Period').agg({'Mao Occurrences': 'sum', 'File Name': 'count'})


period_totals['Average Mentions per Document'] = period_totals['Mao Occurrences'] / period_totals['File Name']

# Plot 
period_totals['Average Mentions per Document'].plot(kind='bar', xlabel='5-Year Periods', ylabel='Average Mentions per Document', title='Average Mentions of Mao per Document per 5-Year Period')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

merged_df['Date'] = pd.to_datetime(merged_df['Date'])

merged_df = merged_df.sort_values(by='Date')

#3 year moving average 
merged_df['3-Year MA'] = merged_df['Mao Occurrences'].rolling(window=3*365, min_periods=1).mean()

plt.figure(figsize=(10, 6))


plt.plot(merged_df['Date'], merged_df['Mao Occurrences'], label='Original Data', marker='o')

#3 year 
plt.plot(merged_df['Date'], merged_df['3-Year MA'], label='3-Year Moving Average', color='red')

plt.title('Mentions of China in SL Documents over Time')
plt.xlabel('Date of Document')
plt.ylabel('mao Occurrences')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt


merged_df['Date'] = pd.to_datetime(merged_df['Date'])

#5-year periods
merged_df['Period'] = merged_df['Date'].dt.year // 5 * 5


period_totals = merged_df.groupby('Period').agg({'Mao Occurrences': 'sum', 'File Name': 'count'})


period_totals['Average Mentions per Document'] = period_totals['Mao Occurrences'] / period_totals['File Name']


period_totals['Start Year'] = period_totals.index
period_totals['End Year'] = period_totals['Start Year'] + 4

# Plot 
period_totals.plot(kind='bar', x='Start Year', y='Average Mentions per Document', xlabel='5-Year Periods', ylabel='Average Mentions per Document', title='Average Mentions of Mao per Document per 5-Year Period')
plt.xticks(rotation=45)
plt.show()
```
